```python
import pandas as pd
```

## The Two Main Pandas Objects
Series: A single column of data (like a list with labels).

DataFrame: A table with rows and columns (like a spreadsheet).



```python
data = pd.Series([10, 20, 30, 40])
```

```python
data
```

```python
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "City": ["New York", "Los Angeles", "Chicago"]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.head(2) #first few rows data
```

```python
df["Name"] #one column as a Series
```

```python
df.loc[1] #Get a specific row by index
```

```python
df[df["Age"] > 30] #Filter Data
```

```python
df["Age Next Year"] = df["Age"] + 1 #Add a New Column
```
